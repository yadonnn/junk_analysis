```python
# 데이터 불러오기
# <=2018 : NTN011, 2019 = ,NTN011  2020 = NTN011
h_16 = pd.read_csv('data-files/2016_4가(생활폐기물관리인원및장비현황).csv')
h_17 = pd.read_csv('data-files/2017_4가(생활폐기물관리인원및장비현황).csv')
h_18 = pd.read_csv('data-files/2018_4가(생활폐기물관리인원및장비현황).csv')
h_19 = pd.read_csv('data-files/2019_다-2-a. (시군구) 관리인원 및 장비현황.csv')
h_20 = pd.read_csv('data-files/2020_3-나-1). (시군구) 관리인원 및 장비현황.csv')

print(h_18.columns)
print(h_19.columns)
print(h_20.columns)
```

```python
def NTN011_preprocessing_before_2019(data,year):
    cols_inc_tot = [c for c in data.columns if '총계' in c]
    data = data.drop(columns=cols_inc_tot)
    data = data.rename(columns={'구분_시도' : '시도', '구분_시군구' : '시군구'})

    data = pd.melt(data, id_vars=['시도','시군구'])
    data['소속기관'], data['인원장비'] = zip(*data['variable'].apply(lambda x: x.split('_')))
    data = data.drop(columns='variable')
    data = data.rename(columns={'value' : '관리인원or장비_대수'})
    data = data[['시도','시군구','소속기관','인원장비','관리인원or장비_대수']]
    data['year'] = year
    return data

df_18 = NTN011_preprocessing_before_2019(h_18, 2018)
df_17 = NTN011_preprocessing_before_2019(h_17, 2017)
df_16 = NTN011_preprocessing_before_2019(h_16, 2016)
```

```python
def NTN011_preprocessing_after_2019(data, year):
    cols_inc_tot = [c for c in data.columns if '총계' in c]
    data = data.drop(columns=cols_inc_tot)

    data = pd.melt(data, id_vars=['시도','시군구'])
    data['소속기관'], data['인원장비'] = zip(*data['variable'].apply(lambda x: x.split('_')))
    data = data.drop(columns='variable')
    data = data.rename(columns={'value' : '관리인원or장비_대수'})
    data = data[['시도','시군구','소속기관','인원장비','관리인원or장비_대수']]
    data['year'] = year
    return data

df_19 = NTN011_preprocessing_after_2019(h_19, 2019)
df_20 = NTN011_preprocessing_after_2019(h_20, 2020)
```

```python
a = pd.DataFrame()
a = pd.concat([df_16, df_17, df_18, df_19, df_20], ignore_index=True)
a.to_csv('data/2016_2020_관리인원_장비현황.csv', index=False)
```
